```python
# Test RippleAdder and Multiplier
K= 14

print('RippleAdder (modular):\n')
for k in range(1,K):
    qc= QuantumCircuit(2*k + 1)
    qc.compose(RippleAdder(k, modular=True), qc.qubits[:], inplace=True)
    qc_transpiled= compiler.transpile(qc, basis_gates=['h', 't', 'tdg', 'x', 'cx'], optimization_level= 0)
    print('Expected CNOTs= ' + str(RippleAdder_CNOTs(k)))
    print('k= %d:  ' %(k) + str(qc_transpiled.count_ops()))

print('RippleAdder (non-modular):\n')
for k in range(1,K):
    qc= QuantumCircuit(2*k + 2)
    qc.compose(RippleAdder(k, modular=False), qc.qubits[:], inplace=True)
    qc_transpiled= compiler.transpile(qc, basis_gates=['h', 't', 'tdg', 'x', 'cx'], optimization_level= 0)
    print('Expected CNOTs= ' + str(RippleAdder_CNOTs(k)+3))
    print('k= %d:  ' %(k) + str(qc_transpiled.count_ops()))

print('\nMultiplier:\n')
for k in range(2,K):
    qc= QuantumCircuit(4*k + 1)
    qc.compose(Multiplier(k), qc.qubits[:], inplace=True)
    qc_transpiled= compiler.transpile(qc, basis_gates=['h', 't', 'tdg', 'x', 'cx'], optimization_level= 0)
    print('Expected CNOTs= ' + str(Multiplier_CNOTs(k)))
    print('k= %d:  ' %(k) + str(qc_transpiled.count_ops()))
```
